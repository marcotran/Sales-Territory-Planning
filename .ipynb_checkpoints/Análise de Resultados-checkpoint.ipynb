### Quantidade de Clientes por Carteira 

```python
df_before = pd.read_csv(path+'original_total_clients_per_salesman.csv')
df_after  = pd.read_csv(path+'optimized_total_clients_per_salesman.csv')
df = pd.merge(df_before, df_after, on='salesman_id', 
              suffixes=('_before', '_after'))[['salesman_id', 'value_before', 'value_after']]

df['salesman_id'] = df.salesman_id.apply(lambda x: str(int(x)))
df['diff'] = df.value_after-df.value_before
df = df.sort_values(['diff'], ascending=True)
df.head()
```

```python
# Initialize the figure with a logarithmic x axis
def plt_total_cli(df, attr):
    f, ax = plt.subplots(figsize=(25, 10))

    plt.title('Alterações de clientes por representante', loc='left', y=1.02)
    sns.barplot(y=attr, x='salesman_id', orient='V', color=sns.color_palette("Blues")[4], order=df.salesman_id, data=df)
    plt.xlabel('Alterações');
    plt.ylabel('Representantes');
    plt.show()
```

```python
plt_total_cli(df, 'value_before')
plt_total_cli(df, 'value_after')
plt_total_cli(df, 'diff')
```

```python
c_latitude  = -16.673960
c_longitude = -49.270990
c_dist      = 5 #km

def map(df, salesman):
    max_avg_ticket = df.client_revenue_avg.max()

    # Build map 
    map_nyc = folium.Map(location=[c_latitude, c_longitude], zoom_start=11, width=840, height=580, 
                        tiles='cartodbpositron')
    # Plot Clients
    for i, row in df.iterrows():
        folium.CircleMarker(location=[row['client_lat'], row['client_lon']], 
                        popup="{}: {}".format(i, row.client_id),
                        color=colors[le.transform([row[salesman]])[0]],
                        fill_color=colors[le.transform([row[salesman]])[0]],
                        radius=(float(row.client_revenue_avg)/float(max_avg_ticket))*8).add_to(map_nyc) 

    # Plot Center Salesman
    df_s = df[['salesman_id','salesman_lat', 'salesman_lon']].drop_duplicates()

    for i, row in df_s.iterrows():
        lat, long = row.salesman_lat, row.salesman_lon
        folium.Marker([lat, long], popup=str(i), icon=folium.Icon(color='#ffff00')).add_to(map_nyc)
        folium.RegularPolygonMarker(
            [lat, long],
            popup=str(row.salesman_id),
            fill_color=colors[le.transform([row.salesman_id])[0]],
            number_of_sides=4,
            radius=7
            ).add_to(map_nyc)
    return map_nyc
#map_nyc.save('mapa.html')
```

### Distância Percorrida

```python
df_before = pd.read_csv(path+'original_dist_per_salesman.csv')
df_after  = pd.read_csv(path+'optimized_dist_per_salesman.csv')
df = pd.merge(df_before, df_after, on='salesman_id', 
              suffixes=('_before', '_after'))[['salesman_id', 'value_before', 'value_after']]

df['salesman_id'] = df.salesman_id.apply(lambda x: str(int(x)))
df['diff'] = df.value_after-df.value_before
df = df.sort_values(['diff'], ascending=True)
df.head()
```

```python
# Initialize the figure with a logarithmic x axis
def plt_dist(df, attr):
    f, ax = plt.subplots(figsize=(25, 10))

    plt.title('Distância entre os clientes por representante', loc='left', y=1.02)
    sns.barplot(y=attr, x='salesman_id', orient='v', color=sns.color_palette("Blues")[4], 
                order=df.salesman_id, data=df)
    plt.xlabel('Distância (KM)');
    plt.ylabel('Representantes');
    plt.show()
    print(np.sum(df[attr]))
```

```python
plt_dist(df, 'value_before')
plt_dist(df, 'value_after')
plt_dist(df, 'diff')
```

### Sustentabilidade

```python
df_before = pd.read_csv(path+'original_sustainability_per_salesman.csv')
df_after  = pd.read_csv(path+'optimized_sustainability_per_salesman.csv')
df = pd.merge(df_before, df_after, on='salesman_id', 
              suffixes=('_before', '_after'))[['salesman_id', 'value_before', 'value_after']]

df['salesman_id'] = df.salesman_id.apply(lambda x: str(int(x)))
df['diff'] = ((df.value_after-df.value_before)/np.fabs(df.value_before))*100
df = df.sort_values(['diff'], ascending=True)
df.head()
```

```python
# Initialize the figure with a logarithmic x axis
def plt_sustentability(df, attr):
    f, ax = plt.subplots(figsize=(25, 10))

    plt.title('Sustentabilidade entre os clientes por representante', loc='left', y=1.02)
    sns.barplot(y=attr, x='salesman_id', orient='v', color=sns.color_palette("Blues")[4], order=df.salesman_id, data=df)
    plt.xlabel('Distância (KM)');
    plt.ylabel('Representantes');
    plt.show()
    print(np.sum(df[attr]))
```

```python
plt_sustentability(df, 'value_before')
plt_sustentability(df, 'value_after')
plt_sustentability(df, 'diff')
```
