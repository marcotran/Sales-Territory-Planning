### Visualização dos dados

Vamos utilizar um corte de no mínimo 40 clientes por carteira e o desvio padrão para tirar os outliers de "clientes" dentro das carteiras restantes

````python
# Total de Clientes por Representante
plt.title("Total de clientes por representante", loc='left')
df.salesman_id_1.value_counts().plot(kind='bar')
````

````python
count_clients = df.salesman_id_1.value_counts()
mean_clients  = count_clients.values.mean()
std_clients   = count_clients.values.std()
````

Retira as carteiras que menos de 40 clientes 

````python
salesman_filtered = []
k = 40
for salesman in count_clients.keys():
    if count_clients[salesman] >= k:
        salesman_filtered.append(salesman)
````

````python
df = df[df.salesman_id_1.isin(salesman_filtered)]
````

````python
# Total de Clientes por Representante
plt.title("Total de clientes por representante", loc='left')
df.salesman_id_1.value_counts().plot(kind='bar')
````

Analisa a média do faturamento dos clientes dentro da carteira, o valor de "revenue_avg" deve seguir uma distribuição normal. Vamos retirar os outliers com +- 3 std da média

````python
# Distribuição dos Valor do Ticket por Representante
plt.title("Distribuição da 'revenue_avg' dos clientes", loc='left')
sns.distplot(df.revenue_avg)
````

````python
def plot_revenue_avg(df):
    # Média do Ticket por Representante
    values = df.pivot_table(index='salesman_id_1', values='revenue_avg', aggfunc=np.mean)
    values.plot(kind='bar')
    plt.title("Média do faturamento por representante (território atual)", loc='left')
    plt.plot()

    # Bloxpot
    plt.figure()
    g = sns.boxplot(x="salesman_id_1", y="revenue_avg", data=df)
    plt.xticks(rotation=90)
    plt.plot()

plot_revenue_avg(df)
````

Retira representantes que tenham a média de 'revenue_avg' da carteira +- 3std da média

````python
rep_revenue_avg = df.pivot_table(index='salesman_id_1', values='revenue_avg', aggfunc=np.mean).revenue_avg
rep_revenue_avg.values.mean(), rep_revenue_avg.values.std()
````

````python
salesman_filtered = []
k = rep_revenue_avg.values.mean()+3*rep_revenue_avg.values.std()
for salesman in rep_revenue_avg.keys():
    if rep_revenue_avg[salesman] <= k:
        salesman_filtered.append(salesman)
````

````python
df = df[df.salesman_id_1.isin(salesman_filtered)]
````

````python
plot_revenue_avg(df)
````

Retira clientes do dataset que tenham -ou+ 3std da média

````python
df.revenue_avg.mean(), df.revenue_avg.std()
````

````python
df = df[(df.revenue_avg > df.revenue_avg.mean()-3*df.revenue_avg.std()) \
        & (df.revenue_avg < df.revenue_avg.mean()+3*df.revenue_avg.std())]
````

````python
# Distribuição dos Valor do Ticket por Representante
plt.title("Distribuição da 'revenue_avg' dos clientes", loc='left')
sns.distplot(df.revenue_avg)
````

````python
plot_revenue_avg(df)
````

### Filtra os clientes pela longitude e latitude, apenas os que estão na circunferência

````python
df.iloc[1371]
````

````python
# Distância em linha reta entra o ponto A e o ponto B
def dist(lat1, long1, lat2, long2):
    return vincenty((lat1, long1), (lat2, long2)).km
````

````python
df['dist'] = df.apply(lambda row: dist(row.lat, row.lon, row.reference_point_lat, row.reference_point_lon), axis=1)
````

````python
sns.boxplot(df.dist)
````

Retira os clientes que estão MUITO longe da origem, provavelmente são outliers.

````python
df = df[df.dist <= df.dist.mean()+df.dist.std()*3]
````

### Plotagem no Mapa

````python
c_latitude  = -16.673960
c_longitude = -49.270990
c_dist      = 5 #km
````

````python
max_avg_ticket = df.revenue_avg.max()

# color
linearColor = folium.LinearColormap(['green', 'yellow', 'red'],
                           vmin=df.salesman_id_1.min(), vmax=df.salesman_id_1.max())

# Build map 
map_nyc = folium.Map(location=[c_latitude, c_longitude], zoom_start=11, width=840, height=580, 
                    tiles='cartodbpositron')
# Plot Clients
for i, row in df.iterrows():
    folium.CircleMarker(location=[row['lat'], row['lon']], 
                    popup="{}: {}".format(i, row.activity),
                    fill_color=linearColor(row['salesman_id_1']),
                    radius=(float(row.revenue_avg)/float(max_avg_ticket))*8).add_to(map_nyc) 
    
# Plot Center Salesman
df_s = df[['salesman_id_1','reference_point_lat', 'reference_point_lon']].drop_duplicates()

for i, row in df_s.iterrows():
    lat, long = row.reference_point_lat, row.reference_point_lon
    folium.Marker([lat, long], popup=str(i), icon=folium.Icon(color='#ffff00')).add_to(map_nyc)
    folium.RegularPolygonMarker(
        [lat, long],
        popup=str(row.salesman_id_1),
        fill_color=linearColor(row.salesman_id_1),
        number_of_sides=4,
        radius=7
        ).add_to(map_nyc)
    
map_nyc.save('mapa.html')
````

Constroi a matriz de prioridade com base na curva ABC, ou seja, constroi de uma forma que o AB não tenha ganho em mudar de representante, deixando a mudança sempre no C

Dessa forma, vamos utilizar uma representação binária já que vamos modificar apenas o C. Os clientes AB colocaremos 1 para os atuais representantes e 0 para os demais... com relação ao C, colocamos 1 para todos pois irá permitir a mudança de qualquer representante

````python
# Clients
df_p = df[['salesman_id_1', 'abc']]
df_p.head()
````

````python
priorities = []

for s in np.unique(df_p.salesman_id_1):
        for c in df_p.index:
            if df_p.loc[c].salesman_id_1 == s:
                priorities.append([s, 1])
            else:
                if df_p.loc[c].abc != 'C':
                    if df_p.loc[c].abc == 'A':
                        priorities.append([c, s, 0.1])
                    else:
                        priorities.append([c, s, 0.3])
                else:
                    priorities.append([c, s, 1])
````

````python
pd_pri = pd.DataFrame(priorities, columns = ['client_id', 'salesman_id', 'priority'])
pd_pri.head()
````

````python
pd_pri.to_csv('input/priority_abc2.csv')
````
